1. Marijuana Use trending by Region

```python
#Read the data
marijuana_data = pd.read_excel('DA 401 Region Dataset.xlsx', sheet_name = 'Marijuana Use')
```

```python
#Generate data for Past Year
past_year_data = marijuana_data[marijuana_data['Frequently of Use'] == 'Past Year ']
past_year_2021 = past_year_data[past_year_data['Year'] == 2021]
```

```python
# Create a bar chart with colors by region
plt.figure(figsize=(5, 6))

# Remove horizontal grid lines
sns.set(style="whitegrid")

custom_palette = {'South': 'royalblue', 'Midwest': 'forestgreen', 'West': 'gold', 'Northest': '#EE3B3B'}
past_year_2021 = past_year_2021.sort_values(by=['Percentage'])


# Create the bar chart
ax = sns.barplot(data=past_year_2021, x='Region', y='Percentage', palette=custom_palette)

plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Past Year Marijuana Use')

plt.ylim(0, 45)

plt.tight_layout()  #the legend does not overlap with the graph
plt.show()
```

```python
ax = sns.lineplot(data=past_year_data, x='Year', y='Percentage', hue='Region', palette=custom_palette, marker='o')

plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Past Year Marijuana Use Trend by Year and Region')

# Move the legend outside of the graph
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')

# Set custom limits for the x-axis (horizontal axis)
plt.xlim(2016.5, 2021.5)  # Adjust the limits as needed

plt.tight_layout()  # Ensure the legend does not overlap with the graph
plt.show()
```
